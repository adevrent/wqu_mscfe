```python
def generate_data(n):
    data_array = []
    
    for i in range(n):
        x = np.random.randn()  # normal distribution with mean = 0, variance = 1
        data_array.append(x)
    
    return data_array  # list
```

```python
# Factorial function with extra argument f(.) function
def f(n):
    if n % 2 == 0:
        return n**2 + 1
    else:
        return n**2

def factorial_modified(n, f=None):
    if f != None:
        n = f(n)
        
    if n == 0 or n ==1 :
        return 1
    else:
        return n * factorial_modified(n-1)
```

```python
factorial_modified(2, f)
```

```python
# Regular factorial
def factorial(n):
    if n == 0 or n ==1 :
        return 1
    else:
        return n * factorial(n-1)
```

```python
# Binomial draw function
from numpy.random import uniform
def mybinomial_rv(n, p):
    """Generates one draw of a binomial random variable X~Bin(n, p)

    Args:
        n (int): Number of trials
        p (float): Probability of success in one trial. 0 <= p <= 1
    """
    assert type(n) == int
    assert 0 <= p and p <= 1
    
    p_array = []  # probability vector for k = 0, 1, ..., n
    for k in range(n):
        p_k = (factorial(n) / (factorial(k) * factorial(n-k))) * p**k * (1-p)**(n-k)
        p_array.append(p_k)
    return p_array
```

```python
# Binomial draw function
from numpy.random import uniform
def binomial_rv(n, p):
    """Generates one draw of a binomial random variable X~Bin(n, p)

    Args:
        n (int): Number of trials
        p (float): Probability of success in one trial. 0 <= p <= 1
    """
    assert type(n) == int
    assert 0 <= p and p <= 1
    
    x = 0  # number of successes
    
    for i in range(n):
        t = uniform(0, 1)
        if t < p:
            x += 1
    
    return x
```

```python
x = binomial_rv(10, 0.5)
x
```

```python
# test with monte carlo simulation
x_array = []
for i in range(100):
    x = binomial_rv(10000, 0.5)
    x_array.append(x)

plt.hist(x_array, bins=20)
```

```python
from numpy.random import uniform

def coinflip_cons(k, n=10, p=0.5):
    counter = 0
    for i in range(n):
        U = uniform(0, 1)
        if U <= p:  # flip is Heads
            counter += 1
        else:  # flip is Tails
            counter = 0
            
        if counter >= k:  # k consecutive Heads achieved.
            print(f"{k} consecutive Heads achieved at round {i + 1}. You are paid $1.")
            return 1
    
    print(f"{k} consecutive Heads not achieved. You are paid nothing.")
    return 0  # k consecutive Heads not achieved.
```

```python
coinflip_cons(3)
```

```python
def coinflip_regular(k, n=10, p=0.5):
    num_success = 0
    for i in range(n):
        U = uniform(0, 1)
        if U <= p:  # flip is Heads
            num_success += 1
    
    if num_success >= k:
        print(f"{num_success} Heads achieved. You are paid $1.")
        return 1
    else:
        print(f"{num_success} Heads achieved. You are paid nothing.")
        return 0
```

```python
coinflip_regular(5)
```

```python
def fib(t):
    if t == 0:
        return 0
    elif t == 1:
        return 1
    else:
        return fib(t - 1) + fib(t - 2)
```

```python
fib(0)  # zero'th fibonacci number
```

```python
[fib(x) for x in range(10)]
```
