```python
# Factorial function with extra argument f(.) function
def f(n):
    if n % 2 == 0:
        return n**2 + 1
    else:
        return n**2

def factorial_modified(n, f=None):
    if f != None:
        n = f(n)
        
    if n == 0 or n ==1 :
        return 1
    else:
        return n * factorial_modified(n-1)
```

```python
# Regular factorial
def factorial(n):
    if n == 0 or n ==1 :
        return 1
    else:
        return n * factorial(n-1)
```

```python
# Binomial draw function
from numpy.random import uniform
def mybinomial_rv(n, p):
    """Generates one draw of a binomial random variable X~Bin(n, p)

    Args:
        n (int): Number of trials
        p (float): Probability of success in one trial. 0 <= p <= 1
    """
    assert type(n) == int
    assert 0 <= p and p <= 1
    
    p_array = []  # probability vector for k = 0, 1, ..., n
    for k in range(n):
        p_k = (factorial(n) / (factorial(k) * factorial(n-k))) * p**k * (1-p)**(n-k)
        p_array.append(p_k)
    return p_array
```

```python
# Binomial draw function
from numpy.random import uniform
def binomial_rv(n, p):
    """Generates one draw of a binomial random variable X~Bin(n, p)

    Args:
        n (int): Number of trials
        p (float): Probability of success in one trial. 0 <= p <= 1
    """
    assert type(n) == int
    assert 0 <= p and p <= 1
    
    x = 0  # number of successes
    
    for i in range(n):
        t = uniform(0, 1)
        if t < p:
            x += 1
    
    return x
```

```python
from numpy.random import uniform

def coinflip_cons(k, n=10, p=0.5):
    counter = 0
    for i in range(n):
        U = uniform(0, 1)
        if U <= p:  # flip is Heads
            counter += 1
        else:  # flip is Tails
            counter = 0
            
        if counter >= k:  # k consecutive Heads achieved.
            print(f"{k} consecutive Heads achieved at round {i + 1}. You are paid $1.")
            return 1
    
    print(f"{k} consecutive Heads not achieved. You are paid nothing.")
    return 0  # k consecutive Heads not achieved.
```

```python
coinflip_cons(3)
```

```python
def coinflip_regular(k, n=10, p=0.5):
    num_success = 0
    for i in range(n):
        U = uniform(0, 1)
        if U <= p:  # flip is Heads
            num_success += 1
    
    if num_success >= k:
        print(f"{num_success} Heads achieved. You are paid $1.")
        return 1
    else:
        print(f"{num_success} Heads achieved. You are paid nothing.")
        return 0
```

```python
coinflip_regular(5)
```

```python
def fib(t):
    """Calculates the fibonacci number at index t, using recursive method.

    Args:
        t (int): index of the fibonacci number. 0 <= t
    """
    if t == 0:
        return 0
    elif t == 1:
        return 1
    else:
        return fib(t - 1) + fib(t - 2)
```

```python
def p(x, coeff):
    """Computes value of the polynomial coeff[0] + coeff[1] * x + coeff[2] * x^2 ... coeff[len(n)] * x^len(n)

    Args:
        x (float): any float number
        coeff (list or tuple): the list which contains the coefficients of powers of x
    """
    polysum = 0
    for i, a in enumerate(coeff):
        polysum += a * x**i
    
    return polysum
```

```python
def f(s):
    """Returns the number of capital letters in a string

    Args:
        s (string): any string
    """
    count = 0
    for char in s:
        if char in "QWERTYUIOPASDFGHJKLZXCVBNM":
            count += 1
    
    return count

f("oOo Foo Bar oOo")
```

```python
# A better method:
def count_uppercase(s):
    return sum(c.isupper() for c in s)
    
count_uppercase("oOo Foo Bar oOo")
```

```python
# Exercise 5.5
def linapprox(f, a, b, n, x):
    """Approximate the value of f(x) where a <= x <= b by piecewise linear interpolation method.

    Args:
        f (function): a function mapping some interval [a, b] to real numbers
        a (float): lowest point of this interval
        b (float): highest point of this interval
        n (int): number of grid points
        x (float): a number such that a <= x <= b
    
    Returns:
        linpol (float): piecewise linear interpolation of f at x, based on n evenly grid points
    """
    # === find first grid point larger than x === #
    point = a
    dx = (b - a) / (n - 1)  # length of dx between grid points
    while point <= x:
        point += dx
    
    df_dx = (f(b) - f(a)) / (b - a)  # slope of the secant line between points b and a
    
    linpol = f(a) + df_dx * (point - a)
    return linpol
```

```python
linapprox(lambda x: x**2, 1, 3, 100, 2)
```

```python
analytic_array = [x**2 for x in np.linspace(0, 1, 100)]
numeric_array = [linapprox(lambda x: x**2, 0, 1, 100, x) for x in np.linspace(0, 1, 100)]

plt.figure()
plt.plot(analytic_array, c="b", label="analytic")
plt.plot(numeric_array, c="r", label="numeric")
plt.legend()

# meh, close enough.
```

Remember the formula for call-put parity: \
\
$C+PV(K)=P+S$

$C - P = S - Ke^{-(rt)}$ \
\
where:
$C=$ Price of the European call option \
$PV(K)= Ke^{-(rt)} =$ Present value of the strike price (K), discounted from the value on the expiration date at the risk-free rate \
$P=$ Price of the European put \
$S=$ Spot price or the current market value of the underlying asset


_____________________

```python
c = callBS
p = putBS
round(c - p, 6)  # price between the call and put option at the same strike and same time until expiration.
```
