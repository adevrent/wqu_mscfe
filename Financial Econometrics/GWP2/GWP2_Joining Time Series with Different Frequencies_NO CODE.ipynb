# Joining Time Series with Different Frequencies

## Definition

Joining time series with different frequencies is the process of combining time series datasets that measured at a varied sampling frequencies such as on a daily, weekly, monthly or quarterly basis. Structured data in time series plays an important role across diverse fields, including finance and economics. 

## Description

In real world, different type of data will be measured at different frequencis. For example, stock closing price data will be updated on daily basis, US Federal Reserve interest rate reported on monthly basis and US GDP growth rate reported on a quarterly basis. Thus, when dealing with these time series data with different frequencies, we can align or synchronize the data to same frequency to facilitate on the data analysis. Resampling involves the change of the time series frequency. There are 2 types of resampling:
1. Upsampling: Increase the data frequency, such as from daily to monthly.
2. Downsampling: Decrease the data frequency, such as from monthly to daily. 

## Demonstration and Diagrams

For example, we have 2 times series data with different frequencies which are US Federal Reserve Rate and US 10-year bond yield.  US Federal Reserve Rate data are reported on monthly basis while US 10-year bond yield are updated on daily basis. For this case, we could could downsample the US 10-year bond yield daily data to monthly data to match with US Federal Reserve Rate data to help to simplify the data analysis process.

## Diagnosis

Below are the methods to identify whether the different time series with different frequencies.
1. Check the data imported from the data source. Most of the data vendor will provide the data's frequency. 
2. Visually check on the graph's time stamp from the data source.
3. Plot the time series data in graph and visually inspect on the time stamp.
4. Use Python to plot the time series data and plot the resampled time series data to make sure ensure all the different time series data are on the same time frequency.

## Damage

Joining time series data of different frequencies could bring some drawbacks also. Resampled data to a lower frequency such as  monthly can sometimes ignore on the short-term fluctuations and the critical patterns that shows in the higher frequency data might not shown in lower frequency data. This can potentially causes misinterpretations and misleading in the data presented
Besides, resampled data can also introduce biased, data missing and lead to inaccurate data analysis. 

## Directions

The challenges can be addressed by: 
1. Choose common aggregation method that can better represent the data.
2. Resampled to lower frequecy time series data by interpolation method which estimate the missing data between observed data. This method can be used if the missing data are expected to be varied smoothly over the time.
3. Use weighted analysis method if missing data is not random by giving more weight to observed data. This method can help to reduce potential biases caused by missing values.
